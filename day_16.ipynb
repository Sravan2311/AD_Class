```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.info()
```

```python
df.rename({'sex':'Gender'},axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df['smoker']=lb.fit_transform(df['smoker'])
df['Gender']=lb.fit_transform(df['Gender'])
df['day']=lb.fit_transform(df['day'])
df['time']=lb.fit_transform(df['time'])
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.heatmap(np.abs(df.corr()),cmap='Blues')
```

```python
sns.heatmap(np.abs(df.corr()),cmap='Reds')
```

```python
sns.heatmap(np.abs(df.corr()),cmap='Blues',annot=True)
```

```python
sns.heatmap(np.abs(df.corr())>0.6,cmap='Blues',annot=True)
```
